## Data Cleaning

```python
df = df.drop(columns = 'Patient Number') #drop patient number, since it is not that much important for analysis 
```

Now, we need to convert these ratings in proper number format

```python
for i in range(df.shape[0]):
    df.loc[i]['Sexual Activity'] = df.loc[i]['Sexual Activity'][0]
    df.loc[i]['Concentration'] = df.loc[i]['Concentration'][0]
    df.loc[i]['Optimisim'] = df.loc[i]['Optimisim'][0]
    print((df.loc[i]['Sexual Activity']),end = " ")
```

So in Suicidal thoughts column we are having repeating values, hence we need to handle it

```python
df['Suicidal thoughts'].replace('YES ', 'YES', inplace=True)
```

```python
df['Suicidal thoughts'].nunique()
```

```python
df['Sadness'].unique()
```

```python
df['Sadness'] = df['Sadness'].map({'Usually':0,'Sometimes':1,'Seldom':2,'Most-Often':3})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in df.columns:
    df[i] = le.fit_transform(df[i])
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc
```

```python
 rfc.fit(x_train, y_train)
```

```python
pred = rfc.score(x_train,y_train)*100
```
